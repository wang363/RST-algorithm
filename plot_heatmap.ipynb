### Weight visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import seaborn as sns

#plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.linewidth'] = 1  # 图框宽度#
plt.rcParams['figure.dpi'] = 300  # plt.show显示分辨率
plt.rcParams['mathtext.fontset'] = 'stix' # 公式的字体可以是这样 formula font
''''''
font = {'family': 'serif',
        'serif': 'Times New Roman', 
        'weight': 'normal',
        #'size': 10
        }
plt.rc('font', **font)

# 找到英文名
def find_name(datapath):
    basename = datapath.split('_data')[0]

    name_mess = pd.read_csv('name_idx.csv', header=None,sep=',')
    name_idx = name_mess.iloc[:,0].values.tolist()
    name_eng = name_mess.iloc[:,1].values.tolist()
    name_index = name_idx.index(basename)
    eng_basename = name_eng[name_index]
    zhongwen = name_idx[name_index]

    return eng_basename, zhongwen 

csv_quanzhong_path = r'乙醇-环己烷#4-1_data_1.csv' # path
eng_basename, zhongwen  = find_name(csv_quanzhong_path)
quanzhong_data = pd.read_csv(csv_quanzhong_path, header=None) # read data
quanzhong = quanzhong_data.values
```

```python
fig, ax = plt.subplots(figsize=(5, 8))
heatmap = sns.heatmap(quanzhong, cmap='BuGn')
colorbar = heatmap.collections[0].colorbar
#colorbar.cax=cbar_ax
set1 = np.min(quanzhong,axis=None)
set2 = set1 + 0.25*(quanzhong.max() - quanzhong.min())
set3 = set1 + 0.5*(quanzhong.max() - quanzhong.min())
set4 = set1 + 0.75*(quanzhong.max() - quanzhong.min())
set5 = np.max(quanzhong)
colorbar.set_ticks([set1, set2, set3, set4, set5])  # Set ticks at 0%, 25%, 50%, 75%, 100%+0.02
colorbar.set_ticklabels(['0%', '25%', '50%', '75%', '100%'])  # Set tick labels as percentages-0.02
new_xlabels = np.arange(1, 15)  # Generating labels from 1 to 14
plt.xticks(np.arange(0.5, 14.5), new_xlabels)  # x-axis 
new_ylabels = np.arange(1, 65, 2)  # Generating labels from 1 to 14
plt.yticks(np.arange(0.5,64.5, 2), new_ylabels)  # x-axis 
#new_ylabels = np.arange(1, 129, 4)  # Generating labels from 1 to 14
#plt.yticks(np.arange(0.5,128.5, 4), new_ylabels)  # x-axis 
#plt.tight_layout(rect=[0, 0, 0.92, 0.95]) # 紧致型，距离右侧0.92，不完全占领
#plt.savefig(zhongwen + '_heatmap.jpg',dpi=300,bbox_inches='tight')#,pad_inches=0.2
#plt.savefig(zhongwen + '_heatmap'+ '.tif', format="tif",dpi=300,bbox_inches='tight')#,pad_inches=0.2
plt.show()
```
